```python
# Modules
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore') #filter some warning messages
import seaborn as sns
```

```python
ca2=np.load('./ca2_detrend.npy',allow_pickle=True)
goa_cc=np.load('./goa_cclme_detrend.npy',allow_pickle=True)
goa_ca=np.load('./goa_ca_detrend.npy',allow_pickle=True)
cc_ca=np.load('./cclme_ca_detrend.npy',allow_pickle=True)
goa_ca_t=np.load('./goa_ca_with_trend.npy',allow_pickle=True)
cc_ca_t=np.load('./cclme_ca_with_trend.npy',allow_pickle=True)
all_freq=[goa_ca_t/10,cc_ca_t/10,goa_ca,cc_ca,goa_cc,ca2]
name=["Hot GOA+dry CA","Hot CC+dry CA","GOA MHW+CA drought","CC MHW+CA drought","GOA MHW leads CC MHW","CA drought (2 years)"]# ,fontsize=17)
```

```python
(all_freq[2][0]-all_freq[2][1]).shape
```

```python
sns.set(style="white", palette="muted", color_codes=True)
plt.figure(figsize=(15,6),dpi=130)
for i in range(2):
    ax1  = plt.subplot(2,2,i+1)
    #ax1 = plt.subplot(figsize=(3,3))
    #ax1 = fig.add_subplot(2,4,i+1)
    bw=1
    nbin=int((max(all_freq[i][0].flatten())-min(all_freq[i][0].flatten()))/bw)

    ax1=sns.distplot(all_freq[i][0].mean(axis=0), bins=nbin,hist=True, kde=True,color='b',label='piControl(30 years)')  #, ax=axes[0, 0])hist=False,
    ax1.tick_params(axis='y', colors='b')
    ax1.tick_params(axis="x",direction="out", pad=.5)
    ax1.set_xlabel('Co-occurrences per decade')
    if (i !=0) :
        ax1.set_ylabel('')
    
    nbin=int((max((all_freq[i][1]-all_freq[i][0]).flatten())-min((all_freq[i][1]-all_freq[i][0]).flatten()))/bw)
    ax2 = ax1.twinx()
    ax2=sns.distplot((all_freq[i][1]-all_freq[i][0]), bins=nbin,hist=True,kde=True,color='r',label='Future(SSP5-8.5,2070-2099)')  #, ax=axes[0, 0])bins=10,
    #ax2.set_xlim([-4,14])
    ax2.tick_params(axis='y', colors='r')
    ax2.set_ylabel('')

    plt.axvline(x=0,linestyle='--',color='gray',zorder=0)
    diff_mean=np.mean(all_freq[i][2])
    #diff_spread=np.std(diff_mean)
    plt.axvline(x=diff_mean.mean(),color='k',alpha=.5,label='MME change',zorder=0)
    #ax2.text(0.4, 0.9, name[i], transform=ax2.transAxes)
    ax2.set_title(name[i],fontweight='bold')

lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
lines = lines_1 + lines_2
labels = labels_1 + labels_2
plt.legend(lines,labels, bbox_to_anchor=(1.4, .75))

for i in range(2,6):
    ax1  = plt.subplot(2,4,i+3)
    #ax1 = plt.subplot(figsize=(3,3))
    #ax1 = fig.add_subplot(2,4,i+1)
    bw=1
    nbin=int((max(all_freq[i][0][0].flatten())-min(all_freq[i][0][0].flatten()))/bw)

    ax1=sns.distplot(all_freq[i][0], bins=nbin,hist=True, kde=True,color='b',label='PI')  #, ax=axes[0, 0])hist=False,np.mean(all_freq[i][0],axis=0)
    ax1.tick_params(axis='y', colors='b')
    ax1.tick_params(axis="x",direction="out", pad=.5)
    ax1.set_xlabel('Co-occurrences per century')
    if (i !=2):
        ax1.set_ylabel('')
        
    nbin=int((max((all_freq[i][1]-all_freq[i][0]).flatten())-min((all_freq[i][1]-all_freq[i][0]).flatten()))/bw)
    ax2 = ax1.twinx()
    ax2=sns.distplot((all_freq[i][1]-all_freq[i][0]), bins=nbin,hist=True,kde=True,color='r',label='SSP585')  #, ax=axes[0, 0])bins=10,
    #ax2.set_xlim([-4,10])
    ax2.tick_params(axis='y', colors='r')
    ax2.set_ylabel('')
    

    plt.axvline(x=0,linestyle='--',color='gray',zorder=0)
    diff_mean=np.mean(all_freq[i][2])
    #diff_spread=np.std(diff_mean)
    plt.axvline(x=diff_mean.mean(),color='k',alpha=.5,label='MME change',zorder=0)
    #ax2.text(0.4, 0.9, name[i], transform=ax2.transAxes)
    ax2.set_title(name[i],fontweight='bold')


plt.subplots_adjust(wspace=.47,hspace=.5)        
#plt.tight_layout(pad=0.5)  
```

```python
ks=[]
mw=[]
from scipy import stats
for i in range(6):  
    x=all_freq[i][1].flatten()
    y=all_freq[i][0].flatten()
    mw.append(stats.mannwhitneyu(x,y,alternative='two-sided'))
    ks.append(stats.kstest( x,y,alternative='two-sided'))
#print(stats.kstest( x,y,alternative='less'))
#print(stats.mannwhitneyu(x,y))#,alternative='greater'))
ks,mw
```

```python
ks=[]
mw=[]
from scipy import stats
for i in range(6):  
    x=all_freq[i][1].flatten()
    y=all_freq[i][0].flatten()
    if i==3 or i ==5:
        al='greater'
        ks.append(stats.kstest( x,y,alternative=al))
        al= 'less'
        mw.append(stats.mannwhitneyu(x,y,alternative=al))
    else :
        al='less'
        ks.append(stats.kstest( x,y,alternative=al))
        al= 'greater'
        mw.append(stats.mannwhitneyu(x,y,alternative=al))

#print(stats.kstest( x,y,alternative='less'))
#print(stats.mannwhitneyu(x,y))#,alternative='greater'))
```
